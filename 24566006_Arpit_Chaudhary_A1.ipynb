**Convert Date of Birth to Age**

Calculating the Age using DOB & Modifing the original dataset and adding Age column to it.




```python
from datetime import datetime

# Load the dataset from CSV
df = pd.read_csv('amazon_prime_users.csv')

# Convert 'Date of Birth' column to datetime format
df['Date of Birth'] = pd.to_datetime(df['Date of Birth'], errors='coerce')

# Get the current year
current_year = pd.Timestamp.now().year

# Calculate the age and add it as a new column
df['Age'] = current_year - df['Date of Birth'].dt.year

# Save the modified dataset to a new CSV file
df.to_csv('modified_amazon_prime_users.csv', index=False)
df.head()
```

# Univariate Analysis  
a. Descriptive Statistics:




```python
# Get summary statistics of numerical columns
df.describe()

# Get summary for categorical columns
df.describe(include=['object'])

```

b. Visualizing Univariate Distributions:

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram for a numerical variable
df['Age'].hist(bins=20)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

c. Subscription Plan Distribution

```python
# Bar chart for a categorical variable
df['Subscription Plan'].value_counts().plot(kind='bar')
plt.title('Subscription Type Distribution')
plt.xlabel('Subscription Type')
plt.ylabel('Count')
plt.show()
```

C. Gender Distribution

```python
# Pie chart for Gender distribution
plt.figure(figsize=(4, 4))
df['Gender'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=['lightblue', 'lightgreen'])
plt.title('Gender Distribution of Amazon Prime Users')
plt.ylabel('')
plt.show()
```

D. Box Plot for Age by Subscription Plan

```python
# Box plot for Age grouped by Subscription Plan
plt.figure(figsize=(10, 6))
sns.boxplot(x='Subscription Plan', y='Age', data=df)
plt.title('Box Plot of Age by Subscription Plan')
plt.xlabel('Subscription Plan')
plt.ylabel('Age')
plt.xticks(rotation=45)
plt.show()
```

E. Stacked Subscription Plan by Gender (Bar Chart)

```python
# Stacked bar chart for Subscription Plan by Gender
plt.figure(figsize=(10, 6))
pd.crosstab(df['Subscription Plan'], df['Gender']).plot(kind='bar', stacked=True, color=['lightblue', 'lightgreen'])
plt.title('Subscription Plan by Gender')
plt.xlabel('Subscription Plan')
plt.ylabel('Count of Users')
plt.xticks(rotation=45)
plt.show()
```

# Multivariate Analysis

a. Correlation Matrix for Numerical Variables:



```python
import seaborn as sns
numeric_df=df.select_dtypes(include=['float64','int64'])
# Correlation matrix
corr_matrix = numeric_df.corr()

# Heatmap for correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

b. Pair Plots:

```python
# Pair plot
sns.pairplot(df)
plt.show()
```

c. Cross Tabulation for Categorical Variables:

```python
# Crosstab for two categorical variables
pd.crosstab(df['Subscription Plan'], df['Payment Information']).plot(kind='bar', stacked=True)
plt.title('Subscription Type vs Payment Information')
plt.show()
```
